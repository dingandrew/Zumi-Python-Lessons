# Loops - for, while

Loops are essentialy pieces of code that allow you to exceute the same code multiple times. 

```python
print("while loop")
i = 0
while i < 10:
    print(i)
    i = i + 1

    
print("For Loop")    
for j in range(0, 10):
    print(j)
```

# For Loops vs While Loops

These two loops perform similar tasks and in theory could be used interchangeably. However each are typically only used for certain tasks. In python there is special syntax that further distinguishes the functionality of for and while.

> While loops are generally used when code needs to be executed until a ceratin condition is met. For example a counnter or  some variable changes.

> For loops are generally used to iterate through an list or some data structure or to execute a certain number of times.



```python
animals = ["cat", "dog", "rat"]

for animal in animals: #this will iterate through animals printing evrything
    print(animal)
    
while "cat" in animals:  #what happened here?
    print("here")
    animals.remove("cat") #try commenting out this line see what happens, press the stop button to stop executing
```

As we saw in the previous example while and for did two very different things the while loop only checked if cat was in animals but the for loop iterated through animals.



```python
#more examples
for i in range(0, 100):
    print(i)
    
while j in range(0, 100): #infinite loop press the stop button!!!!!!
    print(i)
```
